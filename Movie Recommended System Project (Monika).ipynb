```python
import pandas as pd
import numpy as np 
```

now lets get the movie titles

```python
movies = pd.read_csv("tmdb movies.csv")
movies.head()
```

```python
import warnings
warnings.filterwarnings("ignore")
credits = pd.read_csv("credits.csv")
new_credits=credits.filter(['movie_id','title','cast','crew'])
new_credits.head()
```

```python
new_credits.head(5)
```

```python
movies= pd.merge(movies,new_credits,on='title')
movies.head(2)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies.head(2)
```

```python
movies.info()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
movies.head(2)
```

```python
import ast
def convert(obj):
    L =[  ]
    for i in ast.literal_eval(obj):
            L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
import ast
def convert3(obj):
    L =[ ]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter = counter +1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
movies.iloc[0]['overview']
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['title']).toarray()
```

```python
vectors
```

```python
vectors[0]
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import numpy as np
similarity =cosine_similarity(vectors)
```

```python
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title']==movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True , key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('The Dark Knight Rises')
```

```python
recommend('Toy Story')
```

```python
recommend('Spectre')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(ngram_range=(1,2))
tfidf = vectorizer.fit_transform(movies["title"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

#def search(title):
title = "Rises"
query_vec = vectorizer.transform([title])
similarity = cosine_similarity(query_vec,tfidf).flatten()
indices = np.argpartition(similarity,-5)[-5:]
recommend = movies.iloc[indices][::-1]
```

```python
recommend
```
